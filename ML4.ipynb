```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/ML315/refs/heads/main/CreditRisk%20-%20tr.csv')
train.head(3)
```

```python
test = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/ML315/refs/heads/main/CreditRisk%20-%20ts.csv')
test.head(2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train[['Gender','Education','Profession','Income']],train['Default'])
train_acc=knn.score(train[['Gender','Education','Profession','Income']],train['Default'])
print('Train Accuracy',train_acc*100)
```

```python
preds = knn.predict(test[['Gender','Education','Profession','Income']])
preds
```

```python
#print(test['Default'])
```

```python
from sklearn import metrics
test_acc = metrics.accuracy_score(test['Default'],preds)
print('Test Acc',test_acc*100)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test['Default'],preds)
```

```python
from sklearn.metrics import classification_report
print(classification_report(test['Default'],preds))
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(train[['Gender','Education','Profession','Income']],train['Default'])
train_acc=nb.score(train[['Gender','Education','Profession','Income']],train['Default'])
print('Train Accuracy',train_acc*100)
```

```python
preds = knn.predict(test[['Gender','Education','Profession','Income']])
preds
test_acc = metrics.accuracy_score(test['Default'],preds)
print('Test Acc',test_acc*100)
confusion_matrix(test['Default'],preds)
```

```python
print(classification_report(test['Default'],preds))
```
